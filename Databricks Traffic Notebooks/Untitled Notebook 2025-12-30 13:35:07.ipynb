###Extract external location path

```python
# spark.sql("""DESCRIBE EXTERNAL LOCATION `silver`""").select('url')##.display()
bronze_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `bronze`""").select('url').collect()[0]['url']
silver_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `silver`""").select('url').collect()[0]['url']
gold_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `gold`""").select('url').collect()[0]['url']
# print(silver_path)
```

```python
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `bronze` MANAGED LOCATION '{bronze_path}/bronze'""")
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `silver` MANAGED LOCATION '{silver_path}/silver'""")
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `gold` MANAGED LOCATION '{gold_path}/gold'""")
```
