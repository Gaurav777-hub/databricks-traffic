##### 1. Dynamically generate path from ADLSGen2

```python

bronze_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `bronze`""").select('url').collect()[0]['url']
silver_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `silver`""").select('url').collect()[0]['url']
gold_path = spark.sql("""DESCRIBE EXTERNAL LOCATION `gold`""").select('url').collect()[0]['url']
```

##### 2. Create Notebook Parameters using dbutils

```python

dbutils.widgets.text(name="env",defaultValue='',label='Enter the environment name in lower case')
env=dbutils.widgets.get("env")
# print(env)

dbutils.widgets.text(name="schema",defaultValue='',label='Enter the schema name in lower case')
schema=dbutils.widgets.get("schema")
# print(schema)
```

##### 3. Generate schema based on the input parameters

```python

if schema=='silver':
    path = silver_path
elif schema=='gold':
    path = gold_path
elif schema=='bronze':
    path = bronze_path
```

##### 4. Create Function for Schema creation

```python
def create_schema(environment,path):
    print(f'Using {environment}_Catalog')
    spark.sql(f"""USE CATALOG '{environment}_catalog'""")
    print(f'Creating {schema} in {environment}_catalog')
    spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `{schema}` MANAGED LOCATION '{path}/{schema}'""")
```

##### 5. Execute the Function

```python
create_schema(env,path)
```
